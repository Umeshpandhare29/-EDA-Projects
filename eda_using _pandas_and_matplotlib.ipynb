```python
# load dataset using pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('movies_dataset.csv')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
numarical_cols = df[['BudgetUSD', 'US_BoxOfficeUSD', 'Global_BoxOfficeUSD',
       'Opening_Day_SalesUSD', 'One_Week_SalesUSD', 'IMDbRating',
       'RottenTomatoesScore', 'NumVotesIMDb','NumVotesRT']]

for cols in numarical_cols:
    sns.histplot(df[cols],bins=20,kde=True)
    plt.title(f'Distribution of{cols}')
    plt.show()
    
    







```

```python
df.columns


```

```python
categorical_cols = df[['Genre','Country','Director',
       'LeadActor']]


for cols in categorical_cols:
   plt.figure(figsize=(12,8))
   df[cols].value_counts().head(10).plot(kind='bar')
   plt.title(f'top 10 {cols}')
   plt.grid()
   plt.show()
```

```python
x=df['BudgetUSD']
y= df['US_BoxOfficeUSD']

plt.scatter(x,y)
plt.title('box_ofice_income')
plt.grid()
plt.show()
```

```python
x=df['BudgetUSD']
y= df['Global_BoxOfficeUSD']

plt.scatter(x,y)
plt.title('box_ofice_income')
plt.grid()
plt.show()
```

```python
# finding correlation between the columns
corr =[ 'BudgetUSD', 'US_BoxOfficeUSD','Global_BoxOfficeUSD',
       'Opening_Day_SalesUSD', 'One_Week_SalesUSD','IMDbRating']
plt.figure(figsize=(12,6))
sns.heatmap(df[corr].corr(),annot=True)
plt.title('Box_office_budget')
plt.show()
```

from the above observation we can say that there is a strong correlation between budget and revenue but the imdp rating has no relation with budget or revenue so we can say that the revenue of the film is not depend apon the rating 

```python
#
country = df['Country'].value_counts().head(7)
plt.figure(figsize=(10,9))
plt.pie(country,labels=country.index,autopct='%1.1f%%',startangle=90,labeldistance=1.11,pctdistance=0.8)
plt.title('most released  film in countries')
plt.show()
```

from above observation we can say that the usa is the country that release the almost 75%  movies in a years worlwide

```python
plt.scatter(df['BudgetUSD'] / 1e6, df['Global_BoxOfficeUSD'] / 1e6, alpha=0.5, color='purple')
plt.title('Budget vs. Global Box Office Sales')
plt.xlabel('Budget (in Millions USD)')
plt.ylabel('Global Box Office (in Millions USD)')
plt.tight_layout()
```

```python
country_counts = df['Country'].value_counts().sort_values(ascending=False).head(10)
country_counts.plot(kind='bar', color='salmon', edgecolor='black')
plt.title('Top Countries by Number of Movies')
plt.xlabel('Country')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```
