# Exploratory data analysis using seaborn

```python

# comparing the Gender columns
ax = sns.countplot(x ='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

**Sales by gender**

```python
# comparing the gender column with amount column

sales_gen= df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by ='Amount',ascending=False)

sns.barplot(y='Amount',x='Gender',data=sales_gen,hue='Gender')


    
plt.show()
```

**insights:- from above two graph we can say that the count of females is grater than males, also females spend the more  money as compare to males**

**age**

```python
# comparing the age group column with gender
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
for age_count in ax.containers:
    ax.bar_label(age_count)
plt.show()

```

**sales by age Group with gender**

```python
# comparing the age group column with amount
age=df.groupby(['Age Group','Gender'],as_index=False)['Amount'].sum().sort_values(by='Age Group')

sns.barplot(x='Age Group',y='Amount',data=age,hue='Gender' )


plt.show()
```

**sales by marital status**

```python
# # comparing the MArital_status  column with amount to find wchi is doing more shoping maried or unmaried
marital_status=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().reset_index()

sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=marital_status)
plt.show()
```

**insight:** from above two bar plot i observed  that:

the first bar plot tell us the age group 26-35 is spending more amount than other age groups and also detect the female costomers are higher than males costomers.
in second plot we analyze that the unmaried costomers are spending more amount that maried costomers especialy we also find the female costomers is highest than males costomers

so  from this two plot i  can say that the unmaried female costomers  that age between 26-35  has  more contribution  than the males costomers and other categories 


**state**

```python
# comparing the state column with amount to find highest amount comes from which state
plt.figure(figsize=(13,6))

state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='State',y='Amount',data=state)

plt.xticks(rotation=90)
plt.tight_layout()
plt.grid()
plt.show()
```

**insights:**Uttar Pradesh generates the highest revenue among all states, indicating a strong customer base and purchasing power in the region.

**Comparing the occupation column with amount column**

```python
occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(20)
plt.figure(figsize=(10,6))
sns.barplot(x='Occupation',y='Amount',data=occupation)
plt.xticks(rotation=90)
plt.show()
```

**insights:**
Which occupation spends the most?
Which occupation should marketing target?

```python
product=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(20)
plt.figure(figsize=(10,6))
sns.barplot(x='Product_Category',y='Amount',data=product)
plt.xticks(rotation=90)
plt.show()
```

**insights:**
The Food, Clothing, and Electronics categories contribute the highest sales and should be prioritized during festive marketing campaigns.

```python
sns.histplot(x=df['Amount'],data=df,kde=True,bins=30)
plt.show()
```

*insights:* from above histogram we can recognize that the amount between   5000 to 10000 are mostly spended,means most people spend money between 5k to 10k 

```python
plt.figure(figsize=(12,5))
sns.boxplot(y=df['Amount'],data=df)
plt.tight_layout()
plt.show()
```

**insights:**
Several outliers are present, indicating that a small number of customers make unusually large purchases.

```python
sns.pairplot(df[['Age','Orders','Amount']].sample(500))
plt.show()
```

**insights:**
The pairplot indicates that Age, Orders, and Purchase Amount exhibit weak relationships with each other. Most customers are aged between 20 and 40 years, and the majority of purchases fall below ₹12,000. Several high-value purchase outliers are present. Since numerical variables show limited correlation, categorical variables such as Gender, Occupation, State, and Product Category are likely to provide stronger business insights.

```python
# printing heatmap of  analysing correlations  between all numaric column in data
corr=df.corr(numeric_only=True)
plt.figure(figsize=(10,5))
sns.heatmap(corr,annot=True,cmap='viridis')
plt.title('Analysing the correlations of numaric columns')
plt.show()
```

**Key Findings**

 No strong correlation exists between any pair of variables.

**Age and Purchase Amount**

Correlation = 0.031

Very weak positive relationship.

Age has little impact on spending amount.

**Marital Status and Purchase Amount**

Correlation = -0.017

Negligible negative relationship.

Marital status does not significantly affect purchase amount.

**Orders and Purchase Amount**

Correlation = -0.013

Almost no relationship between number of orders and amount spent.

User_ID

User_ID is just an identifier and should not be considered for business insights or predictive modeling.

Project Report Statement

The correlation analysis reveals that none of the numerical features exhibit a strong linear relationship with the purchase amount. All correlation values are close to zero, indicating that factors such as age, marital status, and order count have minimal direct influence on customer spending behavior. Therefore, deeper analysis using categorical variables such as gender, occupation, state, and product category may provide more meaningful business insights.
